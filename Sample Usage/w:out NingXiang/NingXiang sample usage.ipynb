# NingXiang Sample Usage

```python
import NingXiang

import pandas as pd
```

### Import, Prepare data and Read in

```python
train_data = pd.read_csv('../data/curated/modelling/N_Train_1.csv')
```

```python
train_x = train_data.drop(['1'], axis=1)
train_y = train_data['1']
```

```python
ningxiang = NingXiang.NingXiang()
```

```python
ningxiang.read_in_train_data(train_x, train_y)
```

### Setup model type

```python
ningxiang.set_model_type('Regression')
```

### Get Ordered Feature Combination

##### Linear Regression Method

```python
ningxiang.get_lr_based_feature_combinations()
```

##### Random Forest Method

```python
ningxiang.get_rf_based_feature_combinations()
```

### Export NingXiang output

```python
NingXiang.export_ningxiang_output("../models/NingXiang_output_1")
```
