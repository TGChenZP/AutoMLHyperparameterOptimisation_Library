```python
import os

output = pd.DataFrame()
targets = os.listdir(DIR)
targets.sort()

for target in targets:
    if target[-4:] != '.csv':
        continue

    experiment_outcome = pd.read_csv(f'{DIR}/{target}')

    n = len(experiment_outcome)
    n_hits3 = len(experiment_outcome[experiment_outcome['best_found'] == 3])
    n_hits2 = len(experiment_outcome[experiment_outcome['best_found'] == 2])
    n_hits1 = len(experiment_outcome[experiment_outcome['best_found'] == 1])

    perc_hits3 = round((n_hits3)/n, 4)
    perc_hits2 = round((n_hits3+n_hits2)/n, 4)
    perc_hits1 = round((n_hits3+n_hits2+n_hits1)/n, 4)


    if 'batch3' in target:
        hits3_dim_analysis = hits3_dval_analysis = hits3_mean_analysis = hits3_sd_analysis = hits3_cov_analysis = \
            hits2_dim_analysis = hits2_dval_analysis = hits2_mean_analysis = hits2_sd_analysis = hits2_cov_analysis = None

    else:
        hits3_dim_analysis = dict( experiment_outcome[experiment_outcome['best_found'] == 3].value_counts('dim'))
        hits3_dval_analysis = dict( experiment_outcome[experiment_outcome['best_found'] == 3].value_counts('d_val'))
        hits3_mean_analysis = dict( experiment_outcome[experiment_outcome['best_found'] == 3].value_counts('mean_type'))
        hits3_sd_analysis = dict( experiment_outcome[experiment_outcome['best_found'] == 3].value_counts('sd_types'))
        hits3_cov_analysis = dict( experiment_outcome[experiment_outcome['best_found'] == 3].value_counts('cov_type'))

        hits2_dim_analysis = dict( experiment_outcome[(experiment_outcome['best_found'] == 3) | (experiment_outcome['best_found'] == 2)].value_counts('dim'))
        hits2_dval_analysis = dict( experiment_outcome[(experiment_outcome['best_found'] == 3) | (experiment_outcome['best_found'] == 2)].value_counts('d_val'))
        hits2_mean_analysis = dict( experiment_outcome[(experiment_outcome['best_found'] == 3) | (experiment_outcome['best_found'] == 2)].value_counts('mean_type'))
        hits2_sd_analysis = dict( experiment_outcome[(experiment_outcome['best_found'] == 3) | (experiment_outcome['best_found'] == 2)].value_counts('sd_types'))
        hits2_cov_analysis = dict( experiment_outcome[(experiment_outcome['best_found'] == 3) | (experiment_outcome['best_found'] == 2)].value_counts('cov_type'))

    mean_hits3_search = np.round(experiment_outcome[experiment_outcome['best_found'] == 3]['%searched'].mean(), 4)
    med_hits3_search = np.round(experiment_outcome[experiment_outcome['best_found'] == 3]['%searched'].median()  , 4)
    max_hits3_search = np.round(experiment_outcome[experiment_outcome['best_found'] == 3]['%searched'].max()  , 4)

    mean_hits2_search = np.round(experiment_outcome[(experiment_outcome['best_found'] == 3) | (experiment_outcome['best_found'] == 2)]['%searched'].mean()  , 4)
    med_hits2_search = np.round(experiment_outcome[(experiment_outcome['best_found'] == 3) | (experiment_outcome['best_found'] == 2)]['%searched'].median()  , 4)
    max_hits2_search = np.round(experiment_outcome[(experiment_outcome['best_found'] == 3) | (experiment_outcome['best_found'] == 2)]['%searched'].max()  , 4)

    mean_hits1_search = np.round(experiment_outcome[(experiment_outcome['best_found'] == 3) | (experiment_outcome['best_found'] == 2) | (experiment_outcome['best_found'] == 1)]['%searched'].mean()  , 4)
    med_hits1_search = np.round(experiment_outcome[(experiment_outcome['best_found'] == 3) | (experiment_outcome['best_found'] == 2) | (experiment_outcome['best_found'] == 1)]['%searched'].median()  , 4)
    max_hits1_search = np.round(experiment_outcome[(experiment_outcome['best_found'] == 3) | (experiment_outcome['best_found'] == 2) | (experiment_outcome['best_found'] == 1)]['%searched'].max()  , 4)

    mean_hits_search = np.round(experiment_outcome['%searched'].mean()  , 4)
    med_hits_search = np.round(experiment_outcome['%searched'].median()  , 4)
    max_hits_search = np.round(experiment_outcome['%searched'].max()  , 4)


    min_diff0 = np.round(experiment_outcome[experiment_outcome['best_found'] == 0]['diff'].min(), 4)
    mean_diff0 = np.round(experiment_outcome[experiment_outcome['best_found'] == 0]['diff'].mean(), 4)  
    med_diff0 = np.round(experiment_outcome[experiment_outcome['best_found'] == 0]['diff'].median(), 4)  
    max_diff0 = np.round(experiment_outcome[experiment_outcome['best_found'] == 0]['diff'].max(), 4) 

    min_diff = np.round(experiment_outcome[experiment_outcome['best_found'] != 3]['diff'].min(), 4)  
    mean_diff = np.round(experiment_outcome[experiment_outcome['best_found'] != 3]['diff'].mean(), 4)  
    med_diff = np.round(experiment_outcome[experiment_outcome['best_found'] != 3]['diff'].median(), 4)  
    max_diff = np.round(experiment_outcome[experiment_outcome['best_found'] != 3]['diff'].max(), 4)   



    max_time = np.round(experiment_outcome['time'].max(), 4)



    # tmp = pd.DataFrame({'A': [target.split('-')[0][0]], 'B': [target.split('-')[0][1]],
    #         'C': [target.split('-')[0][2]], 'D': [target.split('-')[0][3]],
    #         'E': [target.split('-')[0][4]], 'batch': [target.split('-')[1][5]],  
    #         '% hits 3': [perc_hits3], '% hits 2': [perc_hits2], '% hits 1': [perc_hits1], 
    #         '3 dim': [hits3_dim_analysis], '3 davl': [hits3_dval_analysis], '3 mean': [hits3_mean_analysis], 
    #         '3 sd': [hits3_sd_analysis], '3 cov': [hits3_cov_analysis], 
    #         '2 dim': [hits2_dim_analysis], '2 dval': [hits2_dval_analysis], '2 mean': [hits2_mean_analysis], 
    #         '2 sd': [hits2_sd_analysis], '2 cov': [hits2_cov_analysis], 
    ##         '3 mean search': [mean_hits3_search], '3 med search': [med_hits3_search], '3 max search': [max_hits3_search], 
    # #        '2 mean search': [mean_hits2_search], '2 med search': [med_hits2_search], '2 max search': [max_hits2_search], 
    #  #       '1 mean search': [mean_hits1_search], '1 med search': [med_hits1_search], '1 max search': [max_hits1_search], 
    #         '0 mean search': [mean_hits0_search], '0 med search': [med_hits0_search], '0 max search': [max_hits0_search], 
    #         'mean_diff': [mean_diff], 'med_diff': [med_diff],
    #         'max_time': [max_time]})
    
    tmp = pd.DataFrame({
            # 'A': [target.split('-')[0][0]], 'B': [target.split('-')[0][1]],
            # 'C': [target.split('-')[0][2]], 'D': [target.split('-')[0][3]], 
            'batch': [target.split('-')[1][5]],
            '% hits 3': [perc_hits3], '% hits 2': [perc_hits2], '% hits 1': [perc_hits1], 
            # '3 mean search': [mean_hits3_search], '3 med search': [med_hits3_search], '3 max search': [max_hits3_search], 
            # '2 mean search': [mean_hits2_search], '2 med search': [med_hits2_search], '2 max search': [max_hits2_search], 
            # '1 mean search': [mean_hits1_search], '1 med search': [med_hits1_search], '1 max search': [max_hits1_search], 
            'mean search': [mean_hits_search], 'med search': [med_hits_search], 'max search': [max_hits_search], 
            'min_diff0': [min_diff0], 'mean_diff0': [mean_diff0], 'med_diff0': [med_diff0], 'max_diff0': [max_diff0],
            'min_diff': [min_diff], 'mean_diff': [mean_diff], 'med_diff': [med_diff], 'max_diff': [max_diff],
            'max_time': [max_time]})

    output = output.append(tmp)
```
